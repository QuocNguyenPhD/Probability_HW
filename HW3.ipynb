```python
import numpy as np
import random 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

state_space={0:"a",1:"b",2:"c",3:"d"}

markov=[[0,0.3,0.2,0.5],[0.6,0,0.3,0.1],[0.3,0.4,0,0.3],[0.1,0.7,0.2,0]]
Pa=1/6
Pc=1/3
```

```python
# function that generate Markov chain trajectory
def markov_generator(n): #length of the chain
    
    # choosing initial state
    initial_state=np.random.choice(np.array([0,1,2,3]), size=1 ,p= [Pa,Pa,Pc,Pc])
    
    
    # Generating trajectory
    trajectory=[]
    for i in range(n):
        trajectory.append(state_space[initial_state[0]])
        state= np.random.choice(np.array([0,1,2,3]), size=1 ,p=markov[initial_state[0]])
        initial_state=state
    return(trajectory)   
```

## 3.1

```python
# generate 4000 trajectories

simulate=[]
for i in range(4000):
    simulate.append(markov_generator(500))
```

```python
# X_10
X_10= [X[9] for X in simulate]

# X_100
X_100= [X[99] for X in simulate]

# X_500
X_500= [X[499] for X in simulate]
```

```python
# histogram of 10 step markov chain

duration = ['a', 'b', 'c', 'd']
z = pd.DataFrame({'final state': X_10})

# set up the axis

ax = sns.countplot(z['final state'], order=duration, color='dodgerblue')
ax.set_title('final state at 10 step')
ax.set_xlabel('state')
ax.set_ylabel('frequency')
plt.show()
```

```python
# first time when it each state

T=[]
for j in range(len(simulate)):
    count_a=0
    count_b=0
    count_c=0
    count_d=0
    
    for i in simulate[j]:
       
        if i != 'a':
            count_a= count_a+1  
        else:
            break
            
            # count first time hit a
            
            
    for i in simulate[j]:
        if i !='b':
            count_b= count_b+1
        else:
            break
            # count first time hit b
    
    for i in simulate[j]:        
        if i !='c':
            count_c= count_c+1
        else:
            break
            
            # count first time hit c
            
    for i in simulate[j]:
        if i !='d':
            count_d= count_d+1
        else:
            break
            
            # count first time hit d
        
    T.append([count_a+1,count_b+1,count_c+1, count_d+1])
            
            
            
```

```python
T
```

```python
# histagram plot of first step hit different state

plt.figure(figsize=(15,10), tight_layout="tight")


plt.subplot(2,2,1)
Ta=[X[0] for X in T ]
plt.hist(Ta, bins=20)
plt.title("histogram for First time to hit a in step")
plt.xlabel("step")
plt.ylabel("iteration")




plt.subplot(2,2,2)
Tb=[X[1] for X in T ]
plt.hist(Tb, bins=20)
plt.title("histogram for First time to hit b in step")
plt.xlabel("step")
plt.ylabel("iteration")




plt.subplot(2,2,3)
Tc=[X[2] for X in T ]
plt.hist(Tc, bins=20)
plt.title("histogram for First time to hit c in step")
plt.xlabel("step")
plt.ylabel("iteration")




plt.subplot(2,2,4)
Td=[X[3] for X in T ]
plt.hist(Td, bins=20)
plt.title("histogram for First time to hit d in step")
plt.xlabel("step")
plt.ylabel("iteration")
```

## 3.3

```python
# mean of first time hit different state

Mu_a=np.mean(Ta)
Mu_b=np.mean(Tb)
Mu_c=np.mean(Tc)
Mu_d=np.mean(Td)
print(f"The mean for first time hit a,b,c,d will be: {[Mu_a,Mu_b,Mu_c,Mu_d]}")
```

```python
cov_aa=np.var(Ta)
cov_bb=np.var(Tb)
cov_cc=np.var(Tc)
cov_dd=np.var(Td)
cov_ab=sum((Ta-Mu_a)*(Tb-Mu_b)/3999)
cov_ac=sum((Ta-Mu_a)*(Tc-Mu_c)/3999)
cov_ad=sum((Ta-Mu_a)*(Td-Mu_d)/3999)
cov_bc=sum((Tb-Mu_b)*(Tc-Mu_c)/3999)
cov_bd=sum((Tb-Mu_b)*(Td-Mu_d)/3999)
cov_cd=sum((Tc-Mu_c)*(Td-Mu_d)/3999)
```

```python
# Covariance Matrix of T

cov_matrix=[[cov_aa,cov_ab,cov_ac,cov_ad],[cov_ab,cov_bb,cov_bc,cov_bd],[cov_ac,cov_bc,cov_cc,cov_cd],[cov_ad,cov_bd, cov_cd,cov_dd]]
pd.DataFrame(cov_matrix)
```

```python
#generate N which is the number of abcd in sequence
N=[]

for i in range(len(simulate)):
    
    count=0
    for j in range(len(simulate[0])-3):
        if simulate[i][j] == "a" and simulate[i][j+1]== "b" and simulate[i][j+2]== "c" and simulate[i][j+3]== "d":
            count= count+1
    N.append(count)
```

```python
# calculate probability of N= [0,1,...,20]
Prob_N=[]
for i in range(21):
    count=0
    for j in N:
        if j== i:
            count=count+1
    Prob_N.append(count/4000)
```

```python
Prob_N # probability of N=k for k[0,1,2,3,...,20]
```

```python
# construct confident interval of 95% confident level

pd.DataFrame([np.array(Prob_N)-1.96*np.sqrt(np.array(Prob_N)*(1-np.array(Prob_N))/4000),np.array(Prob_N)+1.96*np.sqrt(np.array(Prob_N)*(1-np.array(Prob_N))/4000)])
```
