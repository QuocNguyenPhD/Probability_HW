```python
import numpy as np
import random 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

state_space={0:"a",1:"b",2:"c",3:"d"}

markov=[[0,0.3,0.2,0.5],[0.6,0,0.3,0.1],[0.3,0.4,0,0.3],[0.1,0.7,0.2,0]]
Pa=1/6
Pc=1/3
```

```python
# function that generate Markov chain trajectory
def markov_generator(n): #length of the chain
    
    # choosing initial state
    initial_state=np.random.choice(np.array([0,1,2,3]), size=1 ,p= [Pa,Pa,Pc,Pc])
    
    
    # Generating trajectory
    trajectory=[]
    for i in range(n):
        trajectory.append(state_space[initial_state[0]])
        state= np.random.choice(np.array([0,1,2,3]), size=1 ,p=markov[initial_state[0]])
        initial_state=state
    return(trajectory)   
```

## 3.1

```python
# generate 4000 trajectories

simulate=[]
for i in range(4000):
    simulate.append(markov_generator(500))
```

```python
# X_10
X_10= [X[9] for X in simulate]

# X_100
X_100= [X[99] for X in simulate]

# X_500
X_500= [X[499] for X in simulate]
```

```python
# histogram of 10 step markov chain

duration = ['a', 'b', 'c', 'd']
z = pd.DataFrame({'final state': X_10})

# set up the axis

ax = sns.countplot(z['final state'], order=duration, color='dodgerblue')
ax.set_title('final state at 10 step')
ax.set_xlabel('state')
ax.set_ylabel('frequency')
plt.show()
```

```python
# histogram of 100 step markov chain


z = pd.DataFrame({'final state': X_100})
ax = sns.countplot(z['final state'], order=duration, color="green")
ax.set_title('final state at 100 step')
ax.set_xlabel('state')
ax.set_ylabel('frequency')
plt.show()
```

```python
# histogram of 500 step markov chain


z = pd.DataFrame({'final state': X_500})
ax = sns.countplot(z['final state'], order=duration, color='red')
ax.set_title('final state at 500 step')
ax.set_xlabel('state')
ax.set_ylabel('frequency')
plt.show()
```
