## 3.1

Since all $S_i$ are poisson process with $\lambda=2$, the gap $S_i-S_{i-1}$ is an exponential distribution with $mean= \displaystyle\frac{1}{\lambda}$

```python
U_10000=[]
T_10000=[]
DT_10000=[]
DU_10000=[]

# create simulation for 10000 time
for i in range(10000):
    # the gap between arrival
    gap=np.random.exponential(1/2, size=3000)

    # time of arrival
    S=[sum(gap[0:(i+1)]) for i in range(3000)]

    # which machine is used
    K= np.random.choice(np.array([0,1]), size=3000 ,p= [1/2,1/2])

    # time arrival to machine 1
    T=[]
    # time arrival to machine 2
    U=[]
    for i in range(3000):
        if K[i]==0:
            T.append(S[i])
        else:
            U.append(S[i])

    # gap between time from machine 1
    DT= [T[i+1]-T[i] for i in range(len(T)-1)]

    # gap between time for machine 2
    DU= [U[i+1]-U[i] for i in range(len(U)-1)]
    
    U_10000.append(U)
    T_10000.append(T)
    DT_10000.append(DT)
    DU_10000.append(DU)
    
```

## 3.2

Gamma distribution density function

$f(x,\alpha,\beta)= \displaystyle\frac{x^{\alpha-1}*\beta^{\alpha}*e^{-\beta*x}}{\Gamma(\alpha)}$

with $mean= \displaystyle\frac{\alpha}{\beta}$ and $var=\displaystyle\frac{\alpha}{\beta^2}$

$\Rightarrow \beta=\displaystyle\frac{mean}{var} \Rightarrow \alpha=\displaystyle\frac{mean^2}{var}$

Hence given mean and std we will find $\alpha, \beta$
