### Task 1

```python
import numpy as np

# Define the parameters
N = 500
r = 3
p1 = 0.18
p2 = 0.42
p3 = 0.40
num_samples = 1000

# Initialize an array to store the simulated vectors
simulated_vectors = []

# Generate random sets and compute vectors
for _ in range(num_samples):
    random_set = np.random.rand(N) #generate a set of 500 uniform randon variables

    # Initialize counters for each voter type
    counts = [0, 0, 0]

    # Determine the voter type for each random number in the set
    for u in random_set:
        if u < p1:
            counts[0] += 1
        elif u < p1 + p2:
            counts[1] += 1
        else:
            counts[2] += 1

    # Append the vector to the list of simulated vectors
    simulated_vectors.append(counts)

# Convert the list of simulated vectors to a NumPy array
simulated_vectors = np.array(simulated_vectors)

# Print the first few simulated vectors
for i in range(1000):
    print(f"V({i + 1}): {simulated_vectors[i]}")
```

### Task 2

```python
# Proportion of voter

g=simulated_vectors/500

# Count the frequency for event A
counts = 0
for i in range(1000):
    if g[i,0]<0.15 and g[i,1]<0.4 and g[i,2]<0.43:
        counts +=1
        
print(f"probability of A occur from the simulationis: {counts/1000}")
```
