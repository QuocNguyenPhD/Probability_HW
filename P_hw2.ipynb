```python
import numpy as np
import matplotlib.pyplot as plt

# Define the values of z
z = np.linspace(0, 5, 100)

# Values of n
n_values = [1, 3, 5, 10, 20]

# Plot the CDF G_n(z) for different values of n
plt.figure(figsize=(10, 6))
for n in n_values:
    Gn = (1 - np.exp(-z))**n
    plt.plot(z, Gn, label=f'n={n}')

plt.xlabel('z')
plt.ylabel('G_n(z)')
plt.legend()
plt.title('CDF G_n(z) for Different Values of n')
plt.grid(True)
plt.show()

# Plot the density function g_n(z) for different values of n
plt.figure(figsize=(10, 6))
for n in n_values:
    gn = n * (1 - np.exp(-z))**(n - 1) * np.exp(-z)
    plt.plot(z, gn, label=f'n={n}')

plt.xlabel('z')
plt.ylabel('g_n(z)')
plt.legend()
plt.title('Density Function g_n(z) for Different Values of n')
plt.grid(True)
plt.show()
```
